# Manual Gradient Descent

Let's do Gradient Descent from Scratch with Mean Sqaured Error as a Loss function and the Gradient Descent Algorithm as the Optimizer for the parameter $w$:

We Have the following Points:

$x=\{0.5, 1.1, 2.5, 3.1\}$ and $y=\{0.4, 1.5, 2.1, 3.5\}$, Let's visualize the Points using Numpy:

```python
x = np.array([.5, 1.1, 2.5, 3.1], dtype='float32')
y = np.array([.4, 1.5, 2.1, 3.5], dtype='float32')
```

```python
# Let's plot the points.
plt.scatter(x, y, color='blue')
plt.xlim(0)
plt.ylim(0)

plt.show()
```

# Problem
Build a Model that can learn from these samples and predict $y$ from a new given $x$.

# Model

We propose the following Model:

$$\hat{y}=wx$$

$w$ is the parameter or weight of the model that should be learned.

# Batch Size: 1 sample

In this first example, we will do point by point training:

### Loss or Objective Function

We Define the Loss on a prediction $x_i$ as:

$$J(w)=(\hat{y}_i - y_i)^2=(wx_i - y_i)^2$$

We have:

$${\partial J(w) \over \partial w}(w)=2x_i(wx_i-y_i)$$

### Optimizer
We Define the Gradient Descent Algorithm as:

\begin{cases}
w_{0}=3 & \text{Initial Random Weight} \\
w_{n} = w_{n-1} - \alpha {\partial J(w) \over \partial w}(w_{n-1}) & \alpha \text{ is the Learning Rate}
\end{cases}

### Training

We will train our model on two epochs.

Let's Pick $\alpha=0.1$ and feed points one-by-one to the optimizer:

#### Epoch 1:

$p_1=(0.5,0.4) \to w_1=3-1.1/10=2.89$ 

$p_2=(1.1,1.5) \to w_2=2.52$ 

$p_3=(2.5,2.1) \to w_3=0.42$ 

$p_4=(3.1,3.5) \to w_4=1.78$

#### Epoch 2:

$p_1=(0.5,0.4) \to w_5=1.73$ 

$p_2=(1.1,1.5) \to w_6=1.64$

$p_3=(2.5,2.1) \to w_7=0.64$ 

$p_4=(3.1,3.5) \to w_8=1.58$

Let's plot the changes in $w_i$:

```python
# Let's plot the points.
rcParams['figure.figsize'] = 17, 10

fig, ax = plt.subplots(nrows=3, ncols=3)

# we plot the points + the weight as a title.
i = 0
for row in ax:
    for col in row:
        col.scatter(x, y, color='blue')
        col.set_title("w_" + str(i))
        i += 1

ax[0][0].plot([0, 3.5], [0, 3.5*3], '--', color='black')

ax[0][1].scatter(x[0], y[0], color='red', s=52)
ax[0][1].plot([0, 3.5], [0, 3.5*2.89], '--', color='red')

ax[0][2].scatter(x[1], y[1], color='red', s=52)
ax[0][2].plot([0, 3.5], [0, 3.5*2.52], '--', color='red')

ax[1][0].scatter(x[2], y[2], color='red', s=52)
ax[1][0].plot([0, 3.5], [0, 3.5*0.42], '--', color='red')

ax[1][1].scatter(x[3], y[3], color='red', s=52)
ax[1][1].plot([0, 3.5], [0, 3.5*1.78], '--', color='red')

ax[1][2].scatter(x[0], y[0], color='green', s=52)
ax[1][2].plot([0, 3.5], [0, 3.5*1.73], '--', color='green')

ax[2][0].scatter(x[1], y[1], color='green', s=52)
ax[2][0].plot([0, 3.5], [0, 3.5*1.64], '--', color='green')

ax[2][1].scatter(x[2], y[2], color='green', s=52)
ax[2][1].plot([0, 3.5], [0, 3.5*0.64], '--', color='green')

ax[2][2].scatter(x[3], y[3], color='green', s=52)
ax[2][2].plot([0, 3.5], [0, 3.5*1.58], '--', color='green')
```

Every point tries to change the weight to minimize its local loss, the way $w$ changes will be different if we consider the global loss at every step, we can calculate the gradient of the total loss or a loss based on a random selection of subpoints (Batches).

# Batch Size: All Points

### Loss or Objective Function

As a Loss function, we will use MSE (Mean Squared Error) over all of the points at each iteration (The loss function value is indepedent from the point you insert):

$$J(w)={1 \over n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2={1 \over 4} \sum_{i=1}^{4}(y_i - wx_i)^2$$

$$J(w)={1 \over 4}(19.07 - 35.9w +17.32w^{2})$$

We have:

$${\partial J(w) \over \partial w}(w)=8.66w - 8.97$$

### Optimizer
We Define the Gradient Descent Algorithm as:

\begin{cases}
w_{0}=3 & \text{Initial Random Weight} \\
w_{n} = w_{n-1} - \alpha {\partial J(w) \over \partial w}(w_{n-1}) & \alpha \text{ is the Learning Rate}
\end{cases}

### Training

We will train our model on 8 epochs (each epoch is one iteration where batch_size=4).

Let's Pick $\alpha=0.1$ and feed points one-by-one to the optimizer:

**Epoch 1**

$w_0 = 3 \to w_1=1.3$ 

**Epoch 2**

$w_1 = 1.3 \to w_2=1.07$ 

**Epoch 3**

$w_2=1.07 \to w_3=1.04$ 

**Epoch 4**

$w_3=1.04 \to w_4=1.036$

**Epoch 5**

$w_4=1.036 \to w_5=1.035$ 

**Epoch 6**

$w_5=1.035 \to w_6=1.035$

**Epoch 7**

$w_6=1.035 \to w_7=1.035$

**Epoch 8**

$w_7=1.035 \to w_8=1.035$

Let's plot the changes in $w_i$:

```python
# Let's plot the points.
rcParams['figure.figsize'] = 17, 10

fig, ax = plt.subplots(nrows=3, ncols=3)

# we plot the points + the weight as a title.
i = 0
for row in ax:
    for col in row:
        col.scatter(x, y, color='blue')
        col.set_title("w_" + str(i))
        i += 1

ax[0][0].plot([0, 3.5], [0, 3.5*3], '--', color='black')
ax[0][1].plot([0, 3.5], [0, 3.5*1.07], '--', color='black')
ax[0][2].plot([0, 3.5], [0, 3.5*1.04], '--', color='black')
ax[1][0].plot([0, 3.5], [0, 3.5*1.036], '--', color='black')
ax[1][1].plot([0, 3.5], [0, 3.5*1.035], '--', color='black')
ax[1][2].plot([0, 3.5], [0, 3.5*1.035], '--', color='black')
ax[2][0].plot([0, 3.5], [0, 3.5*1.035], '--', color='black')
ax[2][1].plot([0, 3.5], [0, 3.5*1.035], '--', color='black')
ax[2][2].plot([0, 3.5], [0, 3.5*1.035], '--', color='black')
```

# Batch Size: 2 Points

In this example, and at each training iteration, we will consider 2 random points, so we need 4 epochs to complete the required 8 iterations:

### Loss or Objective Function

As a Loss function, we will use MSE (Mean Squared Error) over the 2 chosen points at each iteration:

$$J(w)={1 \over 2} \sum_{i=1}^{2}(y_i - wx_i)^2$$

$$J(w)={1 \over 2}[(y_1 - wx_1)^2 + (y_2 - wx_2)^2]$$

We have:

$${\partial J(w) \over \partial w}(w)=w(x_1^2 + x_2^2) - x_1y_1 - x_2y_2$$

### Optimizer
We Define the Gradient Descent Algorithm as:

\begin{cases}
w_{0}=3 & \text{Initial Random Weight} \\
w_{n} = w_{n-1} - \alpha {\partial J(w) \over \partial w}(w_{n-1}) & \alpha \text{ is the Learning Rate}
\end{cases}

### Training

We will train our model on 4 epochs (each epoch consists two iteration where batch_size=2 to loop over all of the points).

Let's Pick $\alpha=0.1$ and feed points one-by-one to the optimizer:

**Epoch 1**

$p_2=(1.1, 1.5),p_4=(3.1,3.5) \to w_1=1.004$

$p_1=(0.5,0.4),p_3=(2.5,2.1) \to w_2=0.8964$ 

**Epoch 2**

$p_1,p_2 \to w_3=0.9505256$

$p_3,p_4 \to w_4=1.0529919984$ 

**Epoch 3**

$p_1,p_3 \to w_5=0.91354719944$

$p_2,p_4 \to w_6=1.17508912965$ 

**Epoch 4**

$p_2,p_3 \to w_7=0.98847263893$

$p_1,p_4 \to w_8=1.11883861695$ 

Let's plot the changes in $w_i$:

```python
# Let's plot the points.
rcParams['figure.figsize'] = 17, 10

fig, ax = plt.subplots(nrows=3, ncols=3)

# we plot the points + the weight as a title.
i = 0
for row in ax:
    for col in row:
        col.scatter(x, y, color='blue')
        col.set_title("w_" + str(i))
        i += 1

ax[0][0].plot([0, 3.5], [0, 3.5*3], '--', color='black')

ax[0][1].scatter([x[1], x[3]], [y[1], y[3]], color='red', s=52)
ax[0][1].plot([0, 3.5], [0, 3.5*1.004], '--', color='red')
ax[0][2].scatter([x[0], x[2]], [y[0], y[2]], color='red', s=52)
ax[0][2].plot([0, 3.5], [0, 3.5*0.896], '--', color='red')

ax[1][0].scatter([x[0], x[1]], [y[0], y[1]], color='green', s=52)
ax[1][0].plot([0, 3.5], [0, 3.5*0.950], '--', color='green')
ax[1][1].scatter([x[2], x[3]], [y[2], y[3]], color='green', s=52)
ax[1][1].plot([0, 3.5], [0, 3.5*1.053], '--', color='green')

ax[1][2].scatter([x[0], x[2]], [y[0], y[2]], color='orange', s=52)
ax[1][2].plot([0, 3.5], [0, 3.5*0.913], '--', color='orange')
ax[2][0].scatter([x[1], x[3]], [y[1], y[3]], color='orange', s=52)
ax[2][0].plot([0, 3.5], [0, 3.5*1.175], '--', color='orange')

ax[2][1].scatter([x[1], x[2]], [y[1], y[2]], color='gray', s=52)
ax[2][1].plot([0, 3.5], [0, 3.5*0.988], '--', color='gray')
ax[2][2].scatter([x[0], x[3]], [y[0], y[3]], color='gray', s=52)
ax[2][2].plot([0, 3.5], [0, 3.5*1.119], '--', color='gray')
```
